Import and Load Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind
```

```python

file_path = '/content/paginasi_dataset.xlsx'
data = pd.read_excel(file_path, parse_dates=['Month'], index_col='Month')
data.head()
```

Split Data into Before and After Pagination

```python

pagination_date = '2019-04-01'

data_before = data[data.index < pagination_date]
data_after = data[data.index >= pagination_date]

data_after = data_after[data_after.index <= '2020-12-31']

data_before.head()
```

```python
data_before.head(), data_after.head
```

Statistical Analysis

```python
from scipy.stats import ttest_ind

# Perform t-test to compare means
t_stat, p_value = ttest_ind(data_before['PPS'], data_after['PPS'], equal_var=False)

print(f'T-statistic: {t_stat}')
print(f'P-value: {p_value}')

# Interpret the results
if p_value < 0.05:
    print('There is a significant difference in PPS before and after pagination.')
else:
    print('There is no significant difference in PPS before and after pagination.')
```

Visualize Data

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(data_before.index, data_before['PPS'], label='Before Pagination', color='blue')
plt.plot(data_after.index, data_after['PPS'], label='After Pagination', color='orange')
plt.axvline(pd.to_datetime(pagination_date), color='red', linestyle='--', label='Pagination Implemented')
plt.title('Pageviews per Session (PPS) Before and After Pagination')
plt.xlabel('Date')
plt.ylabel('Pageviews per Session (PPS)')
plt.legend()
plt.show()
```
